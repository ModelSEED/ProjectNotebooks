# OrthoMCL figures

```python
from pandas import read_csv, set_option, concat
from collections import Counter
from json import load
import ast

set_option("display.max_columns", None)

# load the combined OrthoMCL data and map genomes to SRB organisms
KEGG_module_df = read_csv(r"D:\Users\Andrew Freiburger\Documents\Argonne\Dissemination\Presentations\ASM/module_plot.tsv", sep="\t")
with open("genome_mapping_counts2.json", "r") as jsonIn:  srb_mappings = load(jsonIn)
reversed_srb_mappings = {}
for k,v in srb_mappings.items():
    if "nan" in k:  continue
    pair = ast.literal_eval(k)
    reversed_srb_mappings[pair[1]] = pair[0]
KEGG_module_df.columns = [reversed_srb_mappings[col] if col in reversed_srb_mappings else col for col in KEGG_module_df.columns]
duplicated_organisms = [k for k,v in Counter(KEGG_module_df.columns).items() if v>1]
orgCols = []
for col in KEGG_module_df.columns:
    if col not in orgCols:  orgCols.append(col)
print(duplicated_organisms)
KEGG_module_df = KEGG_module_df.set_index(['module_id', 'module_name', "pathway", "pathway_id"]) 
KEGG_module_df = KEGG_module_df.groupby(by=KEGG_module_df.columns, axis=1).mean()
KEGG_module_df = KEGG_module_df.reset_index()
KEGG_module_df.columns = orgCols
display(KEGG_module_df)

# parse the dataframe for content
metadata_df = KEGG_module_df.loc[:, :"pathway"]
anme_df = concat([metadata_df, KEGG_module_df.loc[:, "ANME-1_GB37.MAG":]], axis=1)
srb_df = concat([metadata_df, KEGG_module_df.loc[:, "Eth-SRB1_sp019314185":"B13-G4_sp014237365a"]], axis=1)

display(metadata_df)
display(anme_df)
display(srb_df)
```

```python
import dataframe_image as dfi

styled_anme_df = anme_df.style.background_gradient()
dfi.export(styled_anme_df, "ANME_orthoMCL.png",max_rows=-1,max_cols=-1)
styled_srb_df = srb_df.style.background_gradient()
dfi.export(styled_srb_df, "SRB_orthoMCL.png",max_rows=-1,max_cols=-1)
```
