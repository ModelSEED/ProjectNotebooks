```python
complex_data = {}
for row_id, d in df_anme_annotation.iterrows():
    genome_id = d['genome_id']
    cluster_id = d['cluster_id']
    ko_id = d['ko']
    rast = d['rast']
    gene_alias = d['gene_alias']
    if genome_id not in complex_data:
        complex_data[genome_id] = {}
    if gene_alias not in complex_data[genome_id]:
        complex_data[genome_id][gene_alias] = {
            'ko': {},
            'rast': {},
            'cluster': {}
        }
    complex_data[genome_id][gene_alias]['ko']['text'] = set() if pd.isna(ko_id) else set(ko_id.split(';'))
    complex_data[genome_id][gene_alias]['rast']['text'] = set() if pd.isna(rast) else set(rast.split(';'))
    complex_data[genome_id][gene_alias]['cluster']['text'] = set() if pd.isna(cluster_id) else set(cluster_id.split(';'))
```

```python
ts = TreeStyle()
ts.show_leaf_name = False
ts.show_scale = False
from ete3 import RectFace
col = 0
for cg in complex_groups:
    for gene_id in cg:
        F = TextFace(gene_id)
        ts.aligned_header.add_face(F, column=col)
        col += 3
arc = False
if arc:
    ts.mode = "c"
    ts.arc_start = -180 # 0 degrees = 3 o'clock
    ts.arc_span = 300
def my_layout(node):
    F = TextFace(node.name, tight_text=False)
    F.margin_right = 5
    add_face_to_node(F, node, column=0, position="branch-top")
    
    if node.name.startswith('p__'):
        node.img_style["bgcolor"] = "white"
    elif node.name.startswith('d__'):
        node.img_style["bgcolor"] = "white"
    elif node.name.startswith('c__'):
        node.img_style["bgcolor"] = "white"
    if node.is_leaf():
        c_data = complex_data.get(node.name, {
            'ko': {},
            'rast': {},
            'cluster': {}
        })
        col = 0
        for cg in complex_groups:
            for gene_id in cg:
                g_data = c_data.get(gene_id, {
            'ko': {},
            'rast': {},
            'cluster': {}
        })
                for otype in ['ko', 'cluster', 'rast']:
                    text = ';'.join(g_data[otype].get('text', set()))
                    if otype == 'rast':
                        if text:
                            text = 'rast'
                    elif otype == 'cluster':
                        text = text.split('/')[-1]
                    F = TextFace(text, tight_text=True)
                    color_code = {
                        'ko': '#30f0e3',
                        'cluster': '#f0ad30',
                        'rast': '#f030da',
                    }
                    if text:
                        F.background.color = color_code[otype]
                    else:
                        F.background.color = "white"

                    F.margin_top = 3
                    F.margin_right = 5
                    F.margin_left = 5
                    F.margin_bottom = 3
                    add_face_to_node(F, node, column=col, position="aligned")
                    col += 1
                #add_face_to_node(TextFace('col 1', tight_text=True), node, column=1, position="aligned")
                #add_face_to_node(TextFace('col 2', tight_text=True), node, column=2, position="aligned")
    #    seq_face = SeqMotifFace(node.sequence, seqtype='aa', seq_format='seq')
    #    add_face_to_node(seq_face, node, column=0, position='aligned')
ts.layout_fn = my_layout
t.render("mytree.png", w=3000, units="mm", tree_style=ts)
from IPython.display import Image
from IPython.core.display import HTML
#t.show(tree_style=ts)
Image(filename='./mytree.png')
```
