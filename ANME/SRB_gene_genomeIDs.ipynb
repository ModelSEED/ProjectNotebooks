```python
# from pandas import DataFrame
genome_genes_mapping = {}
for genome in genomes:
    genome_genes_mapping[genome.id] = []
    for feature in genome.features:
        for alias in feature.aliases:
            genome_genes_mapping[genome.id].append(alias[1])
genes_genome_mapping = {gene:genome for genome, genes in genome_genes_mapping.items() for gene in genes}
# print(genome_genes_mapping[genomes[0].id])
# print(len(genes_genome_mapping))
# print({k: len(v) for k,v in gene_genome_mapping.items()})
# df = DataFrame(gene_genome_mapping)
# display(df)

from json import dump
with open("gene_genome_mapping.json", "w") as out:  dump(genes_genome_mapping, out, indent=3)
```
