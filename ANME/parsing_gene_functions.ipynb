```python
from pandas import read_excel, DataFrame

TmcHmcOhcNhc = read_excel("Syntropic_SRB_Andrew.xlsx", "TmcHmcOhcNhc")
TmcABCD = read_excel("Syntropic_SRB_Andrew.xlsx", "TmcABCD")

nfnAB_flxABCD_etfAB = read_excel("Syntropic_SRB_Andrew.xlsx", "nfnAB_flxABCD_etfAB")
nfnAB_flxABCD_etfAB.columns = nfnAB_flxABCD_etfAB.iloc[0].to_numpy()
nfnAB_flxABCD_etfAB = nfnAB_flxABCD_etfAB.drop([0])
```

# TmcHmcOhcNhc

```python
data = []
for column in TmcHmcOhcNhc.columns:
    if column not in ["Genome", "Unnamed: 13", "Unnamed: 14"]:
        data.extend(list(zip(TmcHmcOhcNhc[column].tolist(), [column]*len(TmcHmcOhcNhc), TmcHmcOhcNhc["Genome"].tolist())))
TmcHmcOhcNhc_df = DataFrame(data)
TmcHmcOhcNhc_df.columns = ["GeneID", "Function", "GenomeID"]
TmcHmcOhcNhc_df
```

# TmcABCD

```python
data = []
for column in TmcABCD.columns:
    if column not in ["Genome", "Narrative objt name"]:
        data.extend(list(zip(TmcABCD[column].tolist(), [column]*len(TmcABCD), TmcABCD["Genome"].tolist())))
TmcABCD_df = DataFrame(data)
TmcABCD_df.columns = ["GeneID", "Function", "GenomeID"]
TmcABCD_df
```

# nfnAB_flxABCD_etfAB

```python
data = []
for column in nfnAB_flxABCD_etfAB.columns:
    if column not in ["Organism", "Narrative objt name"]:
        data.extend(list(zip(nfnAB_flxABCD_etfAB[column].to_numpy(), [column]*len(nfnAB_flxABCD_etfAB), nfnAB_flxABCD_etfAB["Organism"].to_numpy())))
nfnAB_flxABCD_etfAB_df = DataFrame(data)
nfnAB_flxABCD_etfAB_df.columns = ["GeneID", "Function", "GenomeID"]
nfnAB_flxABCD_etfAB_df
```
