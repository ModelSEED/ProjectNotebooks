```python
from pandas import read_excel, DataFrame, set_option, concat

set_option("display.max_rows", None)

TmcHmcOhcNhc = read_excel("Syntropic_SRB_Andrew.xlsx", "TmcHmcOhcNhc")
TmcABCD = read_excel("Syntropic_SRB_Andrew.xlsx", "TmcABCD")

nfnAB_flxABCD_etfAB = read_excel("Syntropic_SRB_Andrew.xlsx", "nfnAB_flxABCD_etfAB")
nfnAB_flxABCD_etfAB.columns = nfnAB_flxABCD_etfAB.iloc[0].to_numpy()
nfnAB_flxABCD_etfAB = nfnAB_flxABCD_etfAB.drop([0])

Hydrogenases = read_excel("Syntropic_SRB_Andrew.xlsx", "Hydrogenases")
Core_Respiratory_chain = read_excel("Syntropic_SRB_Andrew.xlsx", "Core_Respiratory_chain")
HdrA = read_excel("Syntropic_SRB_Andrew.xlsx", "HdrA")
Nitrogenases = read_excel("Syntropic_SRB_Andrew.xlsx", "Nitrogenases")
formate_dehydrogenases = read_excel("Syntropic_SRB_Andrew.xlsx", "formate_dehydrogenases")
```

# TmcHmcOhcNhc

```python
data = []
for column in TmcHmcOhcNhc.columns:
    if column not in ["Genome", "Unnamed: 13", "Unnamed: 14"]:
        data.extend(list(zip(TmcHmcOhcNhc[column].tolist(), [column]*len(TmcHmcOhcNhc), TmcHmcOhcNhc["Genome"].tolist())))
TmcHmcOhcNhc_df = DataFrame(data)
TmcHmcOhcNhc_df.columns = ["GeneID", "Function", "OrganismID"]
TmcHmcOhcNhc_df
```

# formate_dehydrogenases

```python
import re

data = []
for column in formate_dehydrogenases.columns:
    gene_orgnaism_col = formate_dehydrogenases[column].to_numpy()
    gene_orgnaism_col_edited = [re.split(r"(?<=\.1)(\s)", x) for x in gene_orgnaism_col if isinstance(x, str)]
    # print(gene_orgnaism_col_edited)
    for index, pair in enumerate(gene_orgnaism_col_edited):
        # print(pair, end="\t")
        if len(pair) == 3:  gene_orgnaism_col_edited[index] = [pair[0], pair[2], None]
        elif len(pair) == 1:
            new_content = re.split(r"(?=\s[A-Z][a-z])(\s)", pair[0])
            # print(new_content)
            if len(new_content) == 1:
                new_content = re.split(r"(\s)", new_content[0], maxsplit=1)
                # print(new_content)
            gene_orgnaism_col_edited[index] = [new_content[0], new_content[2], None]
        # print(gene_orgnaism_col_edited[index])
    data.extend(gene_orgnaism_col_edited)
# print(data)
    # TODO parse the GeneID from the OrganismID via RegEx that can accommodate the three different GeneID and different OrganismID formats 
#     data.extend(list(zip(formate_dehydrogenases[column].to_numpy(), [column]*len(formate_dehydrogenases), formate_dehydrogenases["Organism"].to_numpy())))
formate_dehydrogenases_df = DataFrame(data)
formate_dehydrogenases_df.columns = ["GeneID", "Function", "OrganismID"]
formate_dehydrogenases_df
```

# Combine the DataFrames

```python
total_df = concat([TmcHmcOhcNhc_df, TmcABCD_df, nfnAB_flxABCD_etfAB_df, Hydrogenases_df, Core_Respiratory_chain_df, 
                   HdrA_df, Nitrogenases_df, formate_dehydrogenases_df], ignore_index=True)
total_df.to_csv("total_GeneID_function.csv")
total_df
```
