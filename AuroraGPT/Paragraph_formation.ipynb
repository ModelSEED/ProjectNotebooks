```python
from chemw import ChemMW

mw = ChemMW(printing=False)

compound_templates = []
test_length = 5
for index, cpd in enumerate(msdb.compounds):
    # print(dir(cpd))
    # print(cpd.annotation)
    compound_reactions = []
    for rxn in cpd.reactions:
        # print(dir(rxn))
        compound_reactions.append(f"""
        """)
    rxn_string = "" if len(compound_reactions) == 0 else f"""{cpd.id} is involved in at least {len(cpd.reactions)} \
reactions: {'; '.join(compound_reactions)}."""
    inchi_string = "" if "InChI" not in cpd.annotation else f""", which is equivalent to {cpd.annotation["InChI"]}"""
    inchiKey_string = "" if "InChIKey" not in cpd.annotation else f""" and has the structural identifier {cpd.annotation["InChIKey"]}"""
    names_string = "" if isinstance(cpd.names, set) else f"""and is also known as {"and ".join([alias for alias in cpd.names])}"""
    metaCyc_id = "" if "MetaCyc" not in cpd.annotation else f""" The MetaCyc ID is {cpd.annotation["MetaCyc"]}."""
    if cpd.formula in ["None", None]:   mw_string = formula_string = ""
    else:
        formula_string = f""" the molecular formula of {cpd.id} is {cpd.formula}"""
        try:  mw_string = f""" with a molecular weight of {cpd.formula_weight if isinstance(cpd.formula_weight, float) else mw.mass(cpd.formula)}"""
        except:  mw_string = ""
    
    compound_template = f"""
        {cpd.name} is a molecule with the ModelSEED ID {cpd.id} {names_string}.{metaCyc_id} At a pH of 7,{formula_string}{mw_string}, \
the charge is {cpd.charge}, and the Delta_G of formation is {cpd.delta_g} with \
an error of {cpd.delta_g_error}. The SMILES molecular structure of {cpd.id} is {cpd.smiles}{inchi_string}{inchiKey_string}. {rxn_string}
    """
    # print(compound_template)
    compound_templates.append(compound_template)
    if len(compound_reactions) != 0:   break
```

```python
compounds_paragraphs = "\n".join(compound_templates)
print(compounds_paragraphs[:1000])

with open("compound_paragraphs.txt", "w") as out:
    out.writelines(compounds_paragraphs)
```
