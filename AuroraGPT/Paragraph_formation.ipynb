```python
def stoich_to_string(stoich_dict):
    string = ""
    reactants, products = {}, {}
    for name,stoich in stoich_dict.items():
        if stoich < 0:   reactants[name] = abs(stoich)
        elif stoich > 0: products[name] = abs(stoich)
        else:   print(f"Error, the {name} has a zero stoich, which is non-sensical")
    string += " and ".join([f"{stoich} {name}" for name, stoich in reactants.items()])
    string += " go to "
    string += " and ".join([f"{stoich} {name}" for name, stoich in products.items()])
    return string
```
