```python
from chemw import ChemMW

mw = ChemMW(printing=False)

compound_templates = []
test_length = 5
for index, cpd in enumerate(msdb.compounds):
    # print(dir(cpd))
    # print(cpd.annotation)
    compound_reactions = []
    for rxn in cpd.reactions:
        # print(dir(rxn))
        compound_reactions.append(f"""
        """)
#     rxn_string = "" if len(compound_reactions) == 0 else f"""{cpd.id} is involved in at least {len(cpd.reactions)} \
# reactions: {'; '.join(compound_reactions)}."""
    inchi_string = "" if "InChI" not in cpd.annotation else f""", which is equivalent to {cpd.annotation["InChI"]}"""
    inchiKey_string = "" if "InChIKey" not in cpd.annotation else f""" and has the structural identifier {cpd.annotation["InChIKey"]}"""
    names_string = "" if not isinstance(cpd.names, set) or any([isinstance(name, float) for name in cpd.names]) else f"""and is also known as {"and ".join([alias for alias in cpd.names])}"""
    metaCyc_id = "" if "MetaCyc" not in cpd.annotation else f""" The MetaCyc ID is {list(cpd.annotation["MetaCyc"])[0]}."""
    
    try:
        mw_string = f""" with a molecular weight of {cpd.formula_weight if isinstance(cpd.formula_weight, float) else mw.mass(cpd.formula)}"""
        formula_string = f""" the molecular formula of {cpd.id} is {cpd.formula}"""
    except:  mw_string = formula_string = ""
        
    metadata_dic = {"ModelSEED ID": cpd.id}
    if metaCyc_id != "":  metadata_dic["MetaCyc"] = list(cpd.annotation["MetaCyc"])[0]
    if hasattr(cpd, "smiles"):  metadata_dic["SMILES"] = cpd.smiles
    if inchi_string != "":  metadata_dic["InChI"] = cpd.annotation["InChI"]
    if inchiKey_string != "":  metadata_dic["InChIKey"] = cpd.annotation["InChIKey"]
    if formula_string != "":  metadata_dic["Formula"] = cpd.formula
    
    compound_template = f"""
        {cpd.name} is a molecule with the ModelSEED ID {cpd.id} {names_string} {metadata_dic}. {metaCyc_id} At a pH of 7,{formula_string}{mw_string}, \
the charge is {cpd.charge}, and the Delta_G of formation is {cpd.delta_g} with \
an error of {cpd.delta_g_error}. The SMILES molecular structure of {cpd.id} is {cpd.smiles}{inchi_string}{inchiKey_string}.
    """
    # print(compound_template)
    compound_templates.append(compound_template)
    if len(compound_reactions) != 0:   break
```
