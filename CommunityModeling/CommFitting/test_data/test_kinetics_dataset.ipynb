```python
def _michaelis_menten(conc, vmax, km):
    return (conc*vmax)/(km+conc)
```

```python
from numpy import linspace, logspace
from pandas import DataFrame
from matplotlib import pyplot
import IPython
# %matplotlib notebook

# define a dataset for an arbitrary circumstance
vmax = 0.4
km = 0.1
data1 = DataFrame()
data1["concentrations"] = logspace(1, 0.03, 100)
data1["growth rates"] = [_michaelis_menten(conc, vmax, km) for conc in data1["concentrations"]]
# render the developed data
fig, ax = pyplot.subplots(dpi=400)
ax.plot(data1.index, data1["concentrations"])
ax.plot(data1.index, data1["growth rates"])
fig.show()

# define a dataset for an arbitrary circumstance #2
vmax = 1
km = 2
data2 = DataFrame()
data2["concentrations"] = logspace(1, 0.1, 100)
data2["growth rates"] = [_michaelis_menten(conc, vmax, km) for conc in data2["concentrations"]]
# render the developed data
fig, ax = pyplot.subplots(dpi=400)
ax.plot(data2.index, data2["concentrations"])
ax.plot(data2.index, data2["growth rates"])
fig.show()
```
