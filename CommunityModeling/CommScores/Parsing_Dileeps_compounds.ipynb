```python
from pandas import read_csv

PMI_Biolog_mops = read_csv("PMI_Biolog_mops.csv")
display(PMI_Biolog_mops)
```

```python
carbon_sources = PMI_Biolog_mops["Carbon source"].unique()
carbon_sources = carbon_sources[1:] ; carbon_sources = carbon_sources[:-1]
print(len(carbon_sources), carbon_sources)
```

```python
from modelseedpy.biochem import from_local
msdb = from_local("../../../ModelSEED/ModelSEEDDatabase")
```

```python
from json import load
import os
with open("../../../ModelSEED/ModelSEEDpy/modelseedpy/data/compoundNames.json", 'r') as cpdNames:   compoundNames = load(cpdNames)

MSIDs = {}
for compound in carbon_sources:
    if compound in compoundNames:
        MSIDs[compound] = compoundNames[compound]
    refined_cpd = "D-"+compound
    # print(refined_cpd)
    if refined_cpd in compoundNames:
        MSIDs[compound] = compoundNames[refined_cpd]
        continue
    refined_cpd = refined_cpd.replace("Acid", "acid")
    # print(refined_cpd)
    if refined_cpd in compoundNames:
        MSIDs[compound] = compoundNames[refined_cpd]
        continue
    refined_cpd = refined_cpd.replace("D-", "")
    # print(refined_cpd)
    if refined_cpd in compoundNames:
        MSIDs[compound] = compoundNames[refined_cpd]
        continue
    refined_cpd = refined_cpd.replace(" acid", "").replace("ic", "ate")
    # print(refined_cpd)
    if refined_cpd in compoundNames:
        MSIDs[compound] = compoundNames[refined_cpd]
        continue
    refined_cpd = refined_cpd.replace("a-", "alpha-")
    # print(refined_cpd)
    if refined_cpd in compoundNames:
        MSIDs[compound] = compoundNames[refined_cpd]
        continue
    refined_cpd = refined_cpd.lower()
    # print(refined_cpd)
    if refined_cpd in compoundNames:
        MSIDs[compound] = compoundNames[refined_cpd]
        continue
    print(compound, "\t", refined_cpd)
    # print("\n\n")
```

```python
print(f"{len(MSIDs)}/{len(carbon_sources)}")
print(MSIDs)
```
