```python
from pandas import read_csv
from json import load
from os import path

# from os import system

# system("ls ~/Documents/CommScores/data")
root_path = "C:/Users/Andrew Freiburger/Documents/Argonne" # "/Users/afreiburger/Documents"
with open(root_path+"/ModelSEED/CommScores/data/compoundNames.json", "r") as jsonIn:
    cpdNameIDs = load(jsonIn)
with open(root_path+"/ModelSEED/CommScores/data/compound_Xrefs.json", "r") as jsonIn:
    cpdXrefs = load(jsonIn)
quantitative_scfa = read_csv(root_path+"/ProjectNotebooks/CommunityModeling/mouse_microbiome/metabolomics data/scfa_quant_results_20220511_PFBBr_GC1315_20220516.csv")
```

```python
tms_metabolomics = read_csv(root_path+"/ProjectNotebooks/CommunityModeling/mouse_microbiome/metabolomics data/GC1315_TMS_Results_Norm.csv")
tms_metabolomics
```

```python
from pandas import set_option
set_option("display.max_columns", None)

unmapped = []
columns = ["SampleName"]
for metName in tms_metabolomics.columns:
    if metName == "SampleName":  continue
    if metName in cpdNameIDs:
        metID = cpdNameIDs[metName]
        if "BiGG" in cpdXrefs[metID]:  columns.append(cpdXrefs[metID]["BiGG"][0])
        else:  print("ID unmapped", metName)  ;  unmapped.append(metName)  ;  columns.append(metName)
    else:  print("name unmapped", metName)  ;  unmapped.append(metName)  ;  columns.append(metName)
print(tms_metabolomics.columns)
print(columns)
tms_metabolomics.columns = columns
tms_metabolomics.to_csv("Bigg_mapped_tms.csv")
tms_metabolomics
```

```python
from pandas import read_csv

megans_mappings = read_csv("BiGG_metabolite_mapping.csv")
```

```python
# display(megans_mappings)
megans_mappings_dict = dict(zip(megans_mappings["Metabolomics ID"], megans_mappings["BiGG ID"]))
print(megans_mappings_dict)
```

```python
from numpy import nan
cols = []
for metName in tms_metabolomics.columns:
    if metName in megans_mappings_dict and megans_mappings_dict[metName] not in ["nan", None, nan]:  cols.append(megans_mappings_dict[metName])
    else:  cols.append(metName)
print(cols)
tms_metabolomics.columns = cols
```
